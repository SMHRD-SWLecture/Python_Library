## 실습 5 - 데이터 시각화 진행

### 실습 5-1
- 제조회사별 hwy 평균, cty 평균 구해서 hwy 평균은 선그래프로, cty의 평균은 막대그래프로 표현

```python
mpg_mean = mpg.groupby('manufacturer').agg(hwy_mean=('hwy', 'mean'),
                                           cty_mean=('cty', 'mean'))
mpg_mean
```

```python
x = mpg_mean.index
y1 = mpg_mean['hwy_mean']
y2 = mpg_mean['cty_mean']
```

```python
plt.figure(figsize=(15,8))
plt.plot(x, y1, label = 'hwy_mean')
plt.bar(x, y2, label = 'cty_mean')
plt.xlabel('manufacturer')
plt.ylabel('mean').set_rotation(0)
plt.legend()
plt.show()
```
