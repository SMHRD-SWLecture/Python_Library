# 7. 최종학력에 따른 월급 평균 확인
- 최종학력 :  EDU_LEVEL

```python
hr_info.groupby('EDU_LEVEL').agg(EDU_SALARY = ('SALARY', 'mean')).sort_values(by = 'EDU_SALARY')
```

# 8. 연차에 따른 월급 평균을 내림차순으로 나열
- SERVICE_YEAR

```python
hr_info.groupby('SERVICE_YEAR').agg(YEAR_SALARY = ('SALARY', 'mean')).sort_values(by = 'YEAR_SALARY', ascending = False)
```

# 9. 최종학력, 성별에 따른 평균 연봉 차이 여부 확인

```python
hr_info.groupby(by = ['EDU_LEVEL', 'GENDER']).agg(SALARY_MEAN = ('SALARY', 'mean'))
```

# 10. 상관계수 분석
- 두 변수 사이의 관계를 수치화하여 보는 방법

## 10-1. 상관계수 파악 방법
- 상관계수 크기(절대값)에 대해
- 0 ~ 0.2 : 관계 없음
- 0.2 ~ 0.4 : 약한 상관관계
- 0.4 ~ 0.6 : 보통 상관관계
- 0.6 ~ 0.8 : 강한 상관관계
- 0.8 ~ 1 : 매우 강한 상관관계



## 10-2. 상관계수 양수/음수
- 양수 : 비례관계
- 음수 : 반비례관계

```python
hr_info[['ORG_CD', 'PERFORMANCE_SCORE', 'SALARY']].groupby(by = 'ORG_CD').mean()
```

# 11. 연도별 입사자 수 확인

```python
hr_info['HIRE_DATE'].str[:4]
```

```python
import matplotlib.pyplot as plt
```

```python
hr_info['HIRE_DATE'] = pd.to_datetime(hr_info['HIRE_DATE'])
hr_info['HIRE_YEAR'] = hr_info['HIRE_DATE'].dt.year
```

```python
yearly_hires = hr_info['HIRE_YEAR'].value_counts().sort_index()
x = yearly_hires.index
y = yearly_hires.values
```

```python
plt.plot(x, y, ls = '--')
plt.xlabel('HIRE_YEAR')
plt.ylabel('HIRE_COUNT')
plt.grid(True, alpha=0.8)
plt.show()
```
