### Matplot Library
- 데이터 시각화를 도와주는 라이브러리

### 데이터 시각화
- 광범위하게 분산된 방대한 양의 자료를 한눈에 볼 수 있도록 도표나 차트 등으로 정리하는 것
- 시각화를 통해서 데이터의 특징을 쉽게 파악이 가능
- 분석 결과를 상대방에게 효과적으로 전달이 가능

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
```

### 선그래프 (line plot) 그리기

```python
y = [2,4,6]
x = [1,2,3]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y)
# show() : 그래프를 보여주는 기능
plt.show()
```

### Line Style 설정하기

```python
y = [2,4,6]
x = [1,2,3]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y, ls = ':')
# show() : 그래프를 보여주는 기능
plt.show()
```

### maker 찍어보기

```python
y = [2,4,6]
x = [1,2,3]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y, marker = '*')
# show() : 그래프를 보여주는 기능
plt.show()
```

### 선의 굵기
- lw

```python
y = [2,4,6]
x = [1,2,3]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y, marker = '*', lw=5)
# show() : 그래프를 보여주는 기능
plt.show()
```

### maker 크기조절
- markersize

```python
y = [2,4,6]
x = [1,2,3]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y, marker = '*', lw=3, markersize = 10)
# show() : 그래프를 보여주는 기능
plt.show()
```

### line color
- color

```python
y = [2,4,6]
x = [1,2,3]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y, marker = '*', lw=3, markersize = 10, color='r', mec='g', mfc='b')
# show() : 그래프를 보여주는 기능
plt.show()
# marker 색깔
# markeredgecolor : 마커 선색깔 : mec
# markerfacecolor : 마커 내부 색깔 : mfc
```

### 그래프 스타일 정리
- linestyle : 선 스타일 (ls)
- linewidth : 선 굵기(lw)
- marker : 마커 종류
- markersize : 마커 크기 조절 (ms)
- color : 선 색깔 (c)
- markerfacecolor : 마커 내부 색깔(mfc)
- markeredgecolor : 마커 선 색깔(mec)
- markeredgewidth : 마커 선 굵기(mew)

### 그래프의 범위를 지정 (x축, y축)
- xlim : x축 범위를 지정
- ylim : y축 범위를 지정

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y, marker = '*', lw=3, markersize = 10, color='r', mec='g', mfc='b')
plt.xlim(-2,10)
plt.ylim(-5,12)
# show() : 그래프를 보여주는 기능 -> 마지막에 사용
plt.show()
```

### 그리드 그리기

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
# plot() : 그래프를 그려주는 함수
plt.plot(x,y, marker = '*', lw=3, markersize = 10, color='r', mec='g', mfc='b')
plt.xlim(-2,10)
plt.ylim(-5,12)
# 틱(tick) 변경
# xticks : x축 틱 설정
# yticks : y축 틱 설정
plt.xticks([0,5,10])
plt.yticks([-2,4,10])
# 그리드 그리기
plt.grid()
# show() : 그래프를 보여주는 기능 -> 마지막에 사용
plt.show()
```

```python
# 여러개 그래프 그리기
# 한 그림안에 여러개 그래프를 그려보자
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.show()
```

### 범례표시
- label

```python
# 폰트변경
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family = font_name)
```

```python
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
# 범례를 보여주는 역활
plt.legend()
plt.show()
```

### 이름설정
- x축의 이름 설정 : xlabel
- y축의 이름 설정 : ylabel
- 그래프의 이름 설정 : title

```python
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
# x,y축의 이름 설정
plt.xlabel('x축')
# set_rotation : y축의 이름의 각도 설정
plt.ylabel('y축').set_rotation(0)
plt.title('제목')
# 범례를 보여주는 역활
plt.legend()
plt.show()
```
